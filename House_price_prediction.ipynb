```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] =(20,10)
```

```python
df1 = pd.read_csv("/content/sample_data/Bengaluru_House_Data.csv")
df1.head()
```

```python
df2 = df1.drop(['society','area_type','availability','balcony'],axis='columns')
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3 = df2.dropna()
df3.isnull().sum()
```

```python
df3['size'].unique()
```

```python
df3['bhk'] = df3['size'].apply(lambda x : int(x.split(" ")[0]))
```

```python
df3.head()
```

To handle no uniform data in dataset we need to do data cleaning

```python
df3['total_sqft'].unique()
```

```python
def iffloat(x):
  try :
    float(x)
  except :
    return False
  return True
```

```python
df3[~df3['total_sqft'].apply(iffloat)].head()
```

Convert ranges of total_sqft and average them using following function

```python
def average(x):
  token = x.split("-")
  if(len(token) == 2):
    return (float(token[0])+float(token[1])/2)
  try:
    return float(x)
  except:
    return None
```

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(average)
df4.head()
```

In real estate price per sqft is very important we create a new column to address the issue

```python
df5 = df4.copy()
df5['price_per_sqft'] = df5['price']*100000/df5['total_sqft']
df5.head()
```

```python
len(df5['location'].unique())
```

Location is a categorical problem it is a high dimensionality problem in this dataset

```python
df5.location = df5.location.apply(lambda x : x.strip())
locationstats = df5.groupby(['location'])['location'].agg('count').sort_values(ascending = False)
locationstats
```

```python
#any location less than 10 is groupes into other
df5.location = df5.location.apply(lambda x : 'other' if x in locationstats[locationstats<=10] else x)
```

OUTLIER REMOVAL


```python
df5.shape
```

```python
#Assume length of each bedroom is atleast 300 practically so
df6 = df5[~(df5.total_sqft/df5.bhk < 300)]
df6.shape
```

Assuming our data set has normal distribution around 69% data points lie between mean and standard deviation and we exclude data sets not lie between these boundaries

```python
df6.price_per_sqft.describe()
```

```python
def removeoutliners(df):
  df_out = pd.DataFrame()
  for key,sub in df.groupby('location'):
    m = np.mean(sub.price_per_sqft)
    st = np.std(sub.price_per_sqft)
    reduced = sub[(sub.price_per_sqft>(m-st)) & (sub.price_per_sqft<=(m+st))]
    df_out = pd.concat([df_out,reduced], ignore_index=True)
  return df_out
```

```python
df7 = removeoutliners(df6)
df7.shape
```

```python
#machine learning model cannot interrupt text column like location we can have to encode them
dummies = pd.get_dummies(df7.location)
dummies.head()
```

```python
df8 = pd.concat([df7,dummies.drop('other', axis = 'columns')], axis = 'columns')
df8.head()
```

```python
df9 = df8.drop(['location','size'],axis = 'columns')
df9.head()
```

```python
X = df9.drop(['price','price_per_sqft'],axis = 'columns')
Y = df9.price
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
qwe = LinearRegression()
qwe.fit(x_train,y_train)
qwe.score(x_test,y_test)
```

```python
def predict_price(location,total_sqft,bath,bhk):
  loc_index = np.where(X.columns==location)[0][0]
  
  x = np.zeros(len(X.columns))
  x[0] = total_sqft
  x[1] = bath
  x[2] = bhk
  if loc_index >=0: 
    x[loc_index] = 1
  
  return qwe.predict([x])[0]
```
